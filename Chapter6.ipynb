```python
# Chapter 6: Working with Tabular Data - Complete Code Examples
# This notebook contains all the code examples from Chapter 6

"""
Required installations before running this notebook:
pip install autogluon.tabular pandas numpy matplotlib seaborn scikit-learn psutil

Optional (for enhanced functionality):
pip install shap  # For model explanations
"""

# =============================================================================
# 6.1 Setting Up AutoGluon and Environment
# =============================================================================

# Installation commands (run in terminal/command prompt)
"""
# Lightweight installation (tabular data only)
pip install autogluon.tabular

# Standard installation (recommended)
pip install autogluon.tabular[all]

# Full installation (all AutoGluon components)
pip install autogluon

# Alternative: Install with conda
conda install -c conda-forge autogluon
"""

# Environment setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')

# Install and import AutoGluon
try:
    from autogluon.tabular import TabularDataset, TabularPredictor
    print("AutoGluon successfully imported!")
except ImportError:
    print("Please install AutoGluon: pip install autogluon.tabular")

# Verify installation and check system
import autogluon.tabular as ag
print(f"AutoGluon Tabular version: {ag.__version__}")

# Note: Available models will be shown after training
print("AutoGluon will automatically select appropriate models based on your data")

# System check
import psutil
import platform

print(f"Platform: {platform.system()} {platform.release()}")
print(f"CPU cores: {psutil.cpu_count()}")
print(f"Available RAM: {psutil.virtual_memory().total / (1024**3):.1f} GB")

# =============================================================================
# 6.2 TabularPredictor Basics - Adult/Census Income Dataset
# =============================================================================

# Load the Adult/Census Income dataset
# This dataset is available from UCI ML Repository
def load_adult_dataset():
    """Load and prepare the Adult dataset"""
    # Column names for the Adult dataset
    columns = [
        'age', 'workclass', 'fnlwgt', 'education', 'education_num',
        'marital_status', 'occupation', 'relationship', 'race', 'sex',
        'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income'
    ]
    
    # Try to load from URL first, then create synthetic data
    try:
        print("Attempting to load Adult dataset from UCI repository...")
        # Load from URL (UCI ML Repository)
        train_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
        test_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.test"
        
        train_data = pd.read_csv(train_url, names=columns, skipinitialspace=True)
        test_data = pd.read_csv(test_url, names=columns, skipinitialspace=True, skiprows=1)
        
        # Combine train and test for our examples
        data = pd.concat([train_data, test_data], ignore_index=True)
        
        # Clean the income column (remove periods and normalize)
        data['income'] = data['income'].str.replace('.', '').str.strip()
        data['income'] = data['income'].map({'>50K': 1, '<=50K': 0})
        
        # Replace '?' with NaN
        data = data.replace('?', np.nan)
        
        print(f"Real Adult dataset loaded successfully: {data.shape}")
        return data
        
    except Exception as e:
        print(f"Could not load real dataset ({e})")
        print("Creating realistic synthetic dataset instead...")
        
        # Create a more realistic synthetic dataset
        np.random.seed(42)
        n_samples = 2000
        
        # Generate correlated features that mimic real Adult dataset patterns
        ages = np.random.normal(39, 13, n_samples)
        ages = np.clip(ages, 17, 90).astype(int)
        
        # Education influences income and hours worked
        education_levels = ['Preschool', '1st-4th', '5th-6th', '7th-8th', '9th', '10th', '11th', '12th',
                           'HS-grad', 'Some-college', 'Assoc-voc', 'Assoc-acdm', 'Bachelors', 'Masters', 'Prof-school', 'Doctorate']
        education_nums = list(range(1, 17))
        education_weights = [0.01, 0.01, 0.01, 0.02, 0.02, 0.02, 0.03, 0.03, 0.32, 0.18, 0.03, 0.03, 0.13, 0.08, 0.05, 0.01]
        
        education = np.random.choice(education_levels, n_samples, p=education_weights)
        education_num = [education_nums[education_levels.index(ed)] for ed in education]
        
        # Generate other features with realistic distributions
        workclass_options = ['Private', 'Self-emp-not-inc', 'Self-emp-inc', 'Federal-gov', 'Local-gov', 'State-gov', 'Without-pay', 'Never-worked']
        workclass_weights = [0.69, 0.08, 0.04, 0.03, 0.06, 0.04, 0.001, 0.001]
        workclass = np.random.choice(workclass_options, n_samples, p=workclass_weights)
        
        sex = np.random.choice(['Male', 'Female'], n_samples, p=[0.67, 0.33])
        
        # Hours per week influenced by sex and education
        base_hours = np.random.normal(40, 12, n_samples)
        hours_per_week = np.clip(base_hours, 1, 99).astype(int)
        
        # Generate income based on education, age, sex, and hours
        income_prob = 0.1  # Base probability
        income_prob += np.array(education_num) * 0.02  # Education effect
        income_prob += (ages - 25) * 0.002  # Age effect
        income_prob += (sex == 'Male') * 0.1  # Gender effect
        income_prob += (hours_per_week - 40) * 0.005  # Hours effect
        income_prob = np.clip(income_prob, 0, 0.8)
        
        income = np.random.binomial(1, income_prob)
        
        data = pd.DataFrame({
            'age': ages,
            'workclass': workclass,
            'fnlwgt': np.random.randint(12285, 1484705, n_samples),  # Census weight
            'education': education,
            'education_num': education_num,
            'marital_status': np.random.choice(['Married-civ-spouse', 'Never-married', 'Divorced', 'Separated', 'Widowed'], n_samples),
            'occupation': np.random.choice(['Prof-specialty', 'Craft-repair', 'Exec-managerial', 'Adm-clerical', 'Sales', 'Other-service'], n_samples),
            'relationship': np.random.choice(['Husband', 'Not-in-family', 'Own-child', 'Unmarried', 'Wife', 'Other-relative'], n_samples),
            'race': np.random.choice(['White', 'Black', 'Asian-Pac-Islander', 'Amer-Indian-Eskimo', 'Other'], n_samples, p=[0.85, 0.1, 0.03, 0.01, 0.01]),
            'sex': sex,
            'capital_gain': np.random.exponential(500, n_samples).astype(int),
            'capital_loss': np.random.exponential(50, n_samples).astype(int),
            'hours_per_week': hours_per_week,
            'native_country': np.random.choice(['United-States', 'Mexico', 'Philippines', 'Germany', 'Canada'], n_samples, p=[0.9, 0.03, 0.02, 0.02, 0.03]),
            'income': income
        })
        
        # Add some missing values to make it realistic
        missing_indices = np.random.choice(n_samples, int(n_samples * 0.05), replace=False)
        data.loc[missing_indices, 'workclass'] = np.nan
        
        missing_indices = np.random.choice(n_samples, int(n_samples * 0.03), replace=False)
        data.loc[missing_indices, 'occupation'] = np.nan
        
        print(f"Synthetic Adult dataset created: {data.shape}")
        return data

# Load the dataset
adult_data = load_adult_dataset()

# Explore the dataset
print("\nDataset Info:")
print(adult_data.info())
print("\nFirst few rows:")
print(adult_data.head())
print("\nTarget distribution:")
print(adult_data['income'].value_counts())

# Basic data exploration
print("\nMissing values:")
missing_summary = adult_data.isnull().sum()
print(missing_summary[missing_summary > 0])

print("\nCategorical columns unique values:")
categorical_cols = adult_data.select_dtypes(include=['object']).columns.tolist()
if len(categorical_cols) > 0:
    for col in categorical_cols:
        print(f"{col}: {adult_data[col].nunique()} unique values")
else:
    print("No categorical columns detected (all features are numeric)")

# Basic TabularPredictor usage - fit and predict
print("\n" + "="*50)
print("BASIC AUTOGLUON MODEL")
print("="*50)

# Split the data
train_data = adult_data.sample(frac=0.8, random_state=42)
test_data = adult_data.drop(train_data.index)

print(f"Training set size: {train_data.shape}")
print(f"Test set size: {test_data.shape}")

# Create and train predictor
predictor = TabularPredictor(
    label='income',
    eval_metric='roc_auc',
    path='./adult_model'
)

# Fit the model
# Note: time_limit can be adjusted based on your available compute resources
# For faster experimentation, use 60-120 seconds
# For better results, use 600-3600 seconds (10-60 minutes)
predictor.fit(train_data, time_limit=120)  # 2 minutes time limit

# Make predictions
predictions = predictor.predict(test_data)
probabilities = predictor.predict_proba(test_data)

print(f"\nPredictions shape: {predictions.shape}")
print(f"Probabilities shape: {probabilities.shape}")

# Calculate and display accuracy
try:
    accuracy = accuracy_score(test_data['income'], predictions)
    print(f"Accuracy: {accuracy:.4f}")
except Exception as e:
    print(f"Could not calculate accuracy: {e}")
    
print(f"\nBasic model training completed successfully!")
print(f"- Dataset size: {adult_data.shape}")
print(f"- Training set: {train_data.shape}")
print(f"- Test set: {test_data.shape}")
print(f"- AutoGluon trained multiple models and selected the best ensemble")

# =============================================================================
# Exploring Predictor Output
# =============================================================================

print("\n" + "="*50)
print("EXPLORING PREDICTOR OUTPUT")
print("="*50)

# Model leaderboard
try:
    leaderboard = predictor.leaderboard(test_data, silent=True)
    print("Model Leaderboard:")
    print(leaderboard)
except Exception as e:
    print(f"Could not generate leaderboard: {e}")
    # Fallback to basic model info
    print("Model information:")
    print(f"Best model: {predictor.model_best}")
    try:
        print(f"Number of models: {len(predictor.model_names())}")
    except:
        print("Model details not available")

# Feature importance
try:
    feature_importance = predictor.feature_importance(test_data)
    print("\nFeature Importance:")
    print(feature_importance.head(10))
except Exception as e:
    print(f"\nCould not compute feature importance: {e}")
    print("This may happen with some model types or configurations")

# Plot feature importance
try:
    if 'feature_importance' in locals() and feature_importance is not None:
        plt.figure(figsize=(10, 6))
        top_features = feature_importance.head(10)
        plt.barh(top_features.index, top_features['importance'])
        plt.title('Top 10 Feature Importance')
        plt.xlabel('Importance')
        plt.tight_layout()
        plt.show()
    else:
        print("Feature importance not available for plotting")
except Exception as e:
    print(f"Could not create plot: {e}")
    if 'feature_importance' in locals() and feature_importance is not None:
        print("Top 10 features by importance:")
        print(feature_importance.head(10))

# Model details
print("\nModel Details:")
print(f"Best model: {predictor.model_best}")
print(f"Models trained: {len(predictor.model_names())}")

# =============================================================================
# Different Prediction Methods
# =============================================================================

print("\n" + "="*50)
print("PREDICTION METHODS")
print("="*50)

# Sample data for prediction
sample_data = test_data.head(5).copy()

# Basic prediction (class labels)
pred_labels = predictor.predict(sample_data)
print("Predicted labels:")
print(pred_labels)

# Prediction probabilities
pred_proba = predictor.predict_proba(sample_data)
print("\nPrediction probabilities:")
print(pred_proba)

# Prediction with specific model (if LightGBM exists)
try:
    available_models = predictor.model_names()
    lgb_model = None
    for model in available_models:
        if 'LightGBM' in model or 'LGB' in model:
            lgb_model = model
            break
    
    if lgb_model:
        specific_model_pred = predictor.predict(sample_data, model=lgb_model)
        print(f"\n{lgb_model} predictions:")
        print(specific_model_pred)
    else:
        print(f"\nNo LightGBM model found. Available models: {available_models[:3]}...")
        if len(available_models) > 0:
            specific_model_pred = predictor.predict(sample_data, model=available_models[0])
            print(f"{available_models[0]} predictions:")
            print(specific_model_pred)
except Exception as e:
    print(f"Could not make model-specific predictions: {e}")

# =============================================================================
# 6.2.1 Binary Classification in Detail
# =============================================================================

print("\n" + "="*50)
print("BINARY CLASSIFICATION DETAILED EXAMPLE")
print("="*50)

# Evaluate the model thoroughly
try:
    eval_metrics = predictor.evaluate(test_data, silent=True)
    print("Evaluation Metrics:")
    for metric, value in eval_metrics.items():
        print(f"{metric}: {value:.4f}")
except Exception as e:
    print(f"Could not evaluate model: {e}")
    # Fallback to basic accuracy if predictions are available
    if 'predictions' in locals():
        try:
            basic_accuracy = accuracy_score(test_data['income'], predictions)
            print(f"Basic accuracy: {basic_accuracy:.4f}")
        except:
            print("Could not compute basic metrics")

# Confusion matrix
try:
    y_true = test_data['income']
    y_pred = predictions
    cm = confusion_matrix(y_true, y_pred)

    try:
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title('Confusion Matrix')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.show()
    except Exception as e:
        print(f"Could not create confusion matrix plot: {e}")
        print("Confusion Matrix:")
        print(f"True Negative: {cm[0,0]}, False Positive: {cm[0,1]}")
        print(f"False Negative: {cm[1,0]}, True Positive: {cm[1,1]}")
        
except Exception as e:
    print(f"Could not compute confusion matrix: {e}")
    print("This might happen if predictions are not available")

# Classification report
try:
    if 'y_true' in locals() and 'y_pred' in locals():
        print("\nClassification Report:")
        print(classification_report(y_true, y_pred))
    else:
        print("\nClassification report not available")
except Exception as e:
    print(f"\nCould not generate classification report: {e}")

# =============================================================================
# 6.2.2 Multi-Class Classification Example
# =============================================================================

print("\n" + "="*50)
print("MULTI-CLASS CLASSIFICATION EXAMPLE")
print("="*50)

# Create a multi-class version using education level
multiclass_data = adult_data.copy()
multiclass_data = multiclass_data.dropna(subset=['education'])

# Use education as target (multi-class)
education_mapping = {
    'Preschool': 'Low', '1st-4th': 'Low', '5th-6th': 'Low', '7th-8th': 'Low',
    '9th': 'Medium', '10th': 'Medium', '11th': 'Medium', '12th': 'Medium',
    'HS-grad': 'Medium', 'Some-college': 'Medium',
    'Assoc-voc': 'High', 'Assoc-acdm': 'High',
    'Bachelors': 'High', 'Masters': 'High', 'Prof-school': 'High', 'Doctorate': 'High'
}

multiclass_data['education_level'] = multiclass_data['education'].map(education_mapping)
multiclass_data = multiclass_data.dropna(subset=['education_level'])

# Remove original education and income columns
features_to_drop = ['education', 'income']
multiclass_data = multiclass_data.drop(columns=features_to_drop)

print(f"Multi-class target distribution:")
print(multiclass_data['education_level'].value_counts())

# Split data
mc_train = multiclass_data.sample(frac=0.8, random_state=42)
mc_test = multiclass_data.drop(mc_train.index)

# Train multi-class predictor
mc_predictor = TabularPredictor(
    label='education_level',
    eval_metric='accuracy',
    path='./multiclass_model'
)

mc_predictor.fit(mc_train, time_limit=120)

# Evaluate
mc_predictions = mc_predictor.predict(mc_test)
mc_accuracy = accuracy_score(mc_test['education_level'], mc_predictions)
print(f"\nMulti-class accuracy: {mc_accuracy:.4f}")

# =============================================================================
# Regression Example
# =============================================================================

print("\n" + "="*50)
print("REGRESSION EXAMPLE")
print("="*50)

# Create a regression dataset (predicting hours worked per week)
regression_data = adult_data.copy()
regression_data = regression_data.dropna(subset=['hours_per_week'])

# Use hours_per_week as target, remove income
reg_features_to_drop = ['income']
regression_data = regression_data.drop(columns=reg_features_to_drop)

# Split data
reg_train = regression_data.sample(frac=0.8, random_state=42)
reg_test = regression_data.drop(reg_train.index)

print(f"Regression target statistics:")
print(reg_train['hours_per_week'].describe())

# Train regression predictor
reg_predictor = TabularPredictor(
    label='hours_per_week',
    problem_type='regression',
    eval_metric='root_mean_squared_error',
    path='./regression_model'
)

reg_predictor.fit(reg_train, time_limit=120)

# Make predictions and evaluate
reg_predictions = reg_predictor.predict(reg_test)
reg_metrics = reg_predictor.evaluate(reg_test, silent=True)

print(f"\nRegression Metrics:")
for metric, value in reg_metrics.items():
    print(f"{metric}: {value:.4f}")

# Plot predictions vs actual
try:
    plt.figure(figsize=(10, 6))
    plt.scatter(reg_test['hours_per_week'], reg_predictions, alpha=0.5)
    plt.plot([reg_test['hours_per_week'].min(), reg_test['hours_per_week'].max()], 
             [reg_test['hours_per_week'].min(), reg_test['hours_per_week'].max()], 'r--')
    plt.xlabel('Actual Hours per Week')
    plt.ylabel('Predicted Hours per Week')
    plt.title('Regression: Predicted vs Actual')
    plt.show()
except Exception as e:
    print(f"Could not create regression plot: {e}")
    print("Regression results summary:")
    print(f"Mean actual: {reg_test['hours_per_week'].mean():.2f}")
    print(f"Mean predicted: {reg_predictions.mean():.2f}")
    print(f"Correlation: {np.corrcoef(reg_test['hours_per_week'], reg_predictions)[0,1]:.3f}")

# =============================================================================
# 6.3 AutoGluon's Automatic Data Processing
# =============================================================================

print("\n" + "="*50)
print("AUTOMATIC DATA PROCESSING DEMONSTRATION")
print("="*50)

# Demonstrate automatic data type detection
print("Automatic Data Type Detection:")
sample_data = adult_data.head(1000).copy()

# Create a predictor to see data type detection
detector_predictor = TabularPredictor(label='income', path='./data_types_demo')

# This will show data type detection in the output
print("Training with automatic data type detection...")
detector_predictor.fit(sample_data, time_limit=60, verbosity=2)

# =============================================================================
# Missing Value Handling Demonstration
# =============================================================================

print("\n" + "="*50)
print("MISSING VALUE HANDLING")
print("="*50)

# Create data with various missing patterns
missing_demo_data = adult_data.sample(1000, random_state=42).copy()

# Introduce different types of missing values
np.random.seed(42)
missing_demo_data.loc[np.random.choice(missing_demo_data.index, 100), 'age'] = np.nan
missing_demo_data.loc[np.random.choice(missing_demo_data.index, 150), 'workclass'] = np.nan
missing_demo_data.loc[np.random.choice(missing_demo_data.index, 50), 'hours_per_week'] = np.nan

print("Missing values by column:")
print(missing_demo_data.isnull().sum())

# Train model with missing values
missing_train, missing_test = train_test_split(missing_demo_data, test_size=0.2, random_state=42)

missing_predictor = TabularPredictor(
    label='income',
    path='./missing_value_demo'
)

missing_predictor.fit(missing_train, time_limit=60)

# Check how it handles missing values in prediction
missing_predictions = missing_predictor.predict(missing_test)
missing_accuracy = accuracy_score(missing_test['income'], missing_predictions)
print(f"Accuracy with missing values: {missing_accuracy:.4f}")

# =============================================================================
# Categorical Encoding Examples
# =============================================================================

print("\n" + "="*50)
print("CATEGORICAL ENCODING DEMONSTRATION")
print("="*50)

# Show different categorical variables and their characteristics
categorical_analysis = {}
categorical_cols = adult_data.select_dtypes(include=['object']).columns.tolist()

if len(categorical_cols) > 0:
    for col in categorical_cols:
        if col in adult_data.columns and col != 'income':
            unique_vals = adult_data[col].nunique()
            most_common_vals = adult_data[col].mode()
            most_common = most_common_vals.iloc[0] if len(most_common_vals) > 0 else 'N/A'
            categorical_analysis[col] = {
                'unique_values': unique_vals,
                'most_common': most_common,
                'encoding_likely': 'one-hot' if unique_vals < 10 else 'label/target'
            }
    
    print("Categorical Variable Analysis:")
    for col, analysis in categorical_analysis.items():
        print(f"{col}: {analysis['unique_values']} unique values, likely encoding: {analysis['encoding_likely']}")
else:
    print("No categorical variables found in the dataset")

# =============================================================================
# Feature Engineering Examples
# =============================================================================

print("\n" + "="*50)
print("FEATURE ENGINEERING EXAMPLES")
print("="*50)

# Demonstrate custom feature engineering
feature_eng_data = adult_data.copy()

# Create datetime features (simulate hire date)
np.random.seed(42)
feature_eng_data['hire_date'] = pd.to_datetime('2020-01-01') + pd.to_timedelta(
    np.random.randint(0, 365*5, len(feature_eng_data)), unit='D'
)

# Extract datetime features
feature_eng_data['hire_year'] = feature_eng_data['hire_date'].dt.year
feature_eng_data['hire_month'] = feature_eng_data['hire_date'].dt.month
feature_eng_data['hire_day_of_week'] = feature_eng_data['hire_date'].dt.dayofweek

# Create interaction features
if 'age' in feature_eng_data.columns and 'hours_per_week' in feature_eng_data.columns:
    feature_eng_data['age_hours_interaction'] = feature_eng_data['age'] * feature_eng_data['hours_per_week']

# Create derived features
if 'capital_gain' in feature_eng_data.columns and 'capital_loss' in feature_eng_data.columns:
    feature_eng_data['net_capital'] = feature_eng_data['capital_gain'] - feature_eng_data['capital_loss']
    feature_eng_data['has_capital_gain'] = (feature_eng_data['capital_gain'] > 0).astype(int)

print("New features created:")
new_features = ['hire_year', 'hire_month', 'hire_day_of_week', 'age_hours_interaction', 'net_capital', 'has_capital_gain']
for feature in new_features:
    if feature in feature_eng_data.columns:
        print(f"- {feature}: {feature_eng_data[feature].dtype}")

# =============================================================================
# 6.4 Advanced Customization Examples
# =============================================================================

print("\n" + "="*50)
print("ADVANCED CUSTOMIZATION FOR TABULAR MODELS")
print("="*50)

# Model selection and exclusion
print("Model Selection Customization:")

# Include only specific models
custom_predictor_1 = TabularPredictor(
    label='income',
    path='./custom_models_1'
)

# Specify which models to include
custom_predictor_1.fit(
    train_data,
    included_model_types=['GBM', 'RF', 'XT'],  # Only gradient boosting, random forest, extra trees
    time_limit=60
)

print("Models in custom predictor 1:")
print(custom_predictor_1.model_names)

# Exclude specific models
custom_predictor_2 = TabularPredictor(
    label='income',
    path='./custom_models_2'
)

custom_predictor_2.fit(
    train_data,
    excluded_model_types=['NN_TORCH'],  # Exclude neural networks
    time_limit=60
)

print("Models in custom predictor 2:")
print(custom_predictor_2.model_names)

# =============================================================================
# Hyperparameter Customization
# =============================================================================

print("\n" + "="*30)
print("HYPERPARAMETER CUSTOMIZATION")
print("="*30)

# Custom hyperparameters for specific models
hyperparameters = {
    'GBM': [
        {'num_boost_round': 100, 'num_leaves': 26},
        {'num_boost_round': 200, 'num_leaves': 30},
    ],
    'RF': [
        {'n_estimators': 100, 'max_depth': 15},
        {'n_estimators': 200, 'max_depth': 20},
    ],
}

custom_hyperparams_predictor = TabularPredictor(
    label='income',
    path='./custom_hyperparams'
)

custom_hyperparams_predictor.fit(
    train_data,
    hyperparameters=hyperparameters,
    time_limit=120
)

print("Custom hyperparameters leaderboard:")
custom_leaderboard = custom_hyperparams_predictor.leaderboard(test_data, silent=True)
print(custom_leaderboard.head())

# =============================================================================
# Advanced Ensemble Configuration
# =============================================================================

print("\n" + "="*30)
print("ENSEMBLE CUSTOMIZATION")
print("="*30)

# Configure ensemble settings
ensemble_predictor = TabularPredictor(
    label='income',
    path='./ensemble_config'
)

# Fit with specific ensemble configuration
ensemble_predictor.fit(
    train_data,
    num_bag_folds=3,      # Number of bagging folds
    num_bag_sets=2,       # Number of bagging sets
    num_stack_levels=2,   # Number of stacking levels
    time_limit=120
)

print("Ensemble model performance:")
ensemble_leaderboard = ensemble_predictor.leaderboard(test_data, silent=True)
print(ensemble_leaderboard.head())

# =============================================================================
# Custom Feature Engineering
# =============================================================================

print("\n" + "="*30)
print("CUSTOM FEATURE ENGINEERING")
print("="*30)

def custom_feature_engineering(df):
    """Custom feature engineering function"""
    df = df.copy()
    
    # Age groups
    if 'age' in df.columns:
        df['age_group'] = pd.cut(df['age'], 
                                bins=[0, 25, 35, 50, 65, 100], 
                                labels=['young', 'adult', 'middle', 'senior', 'elderly'])
    
    # Income potential (based on education and age)
    if 'education_num' in df.columns and 'age' in df.columns:
        df['income_potential'] = df['education_num'] * np.log(df['age'] + 1)
    
    # Work intensity
    if 'hours_per_week' in df.columns:
        df['work_intensity'] = pd.cut(df['hours_per_week'],
                                     bins=[0, 30, 40, 50, 100],
                                     labels=['part_time', 'full_time', 'overtime', 'extreme'])
    
    return df

# Apply custom feature engineering
train_custom_features = custom_feature_engineering(train_data)
test_custom_features = custom_feature_engineering(test_data)

# Train with custom features
custom_features_predictor = TabularPredictor(
    label='income',
    path='./custom_features'
)

custom_features_predictor.fit(train_custom_features, time_limit=120)

# Evaluate
custom_features_pred = custom_features_predictor.predict(test_custom_features)
custom_features_accuracy = accuracy_score(test_custom_features['income'], custom_features_pred)
print(f"Accuracy with custom features: {custom_features_accuracy:.4f}")

# =============================================================================
# Model Interpretability and Debugging
# =============================================================================

print("\n" + "="*30)
print("MODEL INTERPRETABILITY")
print("="*30)

# Feature importance analysis
feature_importance = predictor.feature_importance(test_data)
print("Top 10 Most Important Features:")
print(feature_importance.head(10))

# SHAP values (if available)
try:
    import shap
    
    # Get SHAP values for a sample
    sample_for_shap = test_data.head(100)
    
    # Note: This might not work with all AutoGluon models
    # It's primarily for demonstration
    print("\nSHAP analysis would go here...")
    print("(SHAP integration depends on the specific models used)")
    
except ImportError:
    print("SHAP not available. Install with: pip install shap")

# Model-specific analysis
best_model = predictor.model_best
print(f"\nBest model: {best_model}")

# Performance by model type
print("\nPerformance by model type:")
leaderboard_full = predictor.leaderboard(test_data, silent=True)
model_types = leaderboard_full['model'].str.extract(r'([A-Z]+)')[0]
leaderboard_full['model_type'] = model_types
type_performance = leaderboard_full.groupby('model_type')['score_val'].agg(['mean', 'std', 'count'])
print(type_performance)

# =============================================================================
# 6.5 PROJECT: TITANIC SURVIVAL PREDICTION
# =============================================================================

print("\n" + "="*50)
print("PROJECT: TITANIC SURVIVAL PREDICTION")
print("="*50)

# Load Titanic dataset
def load_titanic_dataset():
    """Load and prepare the Titanic dataset"""
    try:
        # Try to load from seaborn first
        import seaborn as sns
        titanic = sns.load_dataset('titanic')
        
        # Clean up the dataset for our use
        # Map boolean survived to int if needed
        if titanic['survived'].dtype == bool:
            titanic['survived'] = titanic['survived'].astype(int)
            
        print("Loaded Titanic dataset from seaborn")
        return titanic
        
    except Exception as e:
        print(f"Could not load from seaborn ({e})")
        print("Creating synthetic Titanic dataset...")
        
        # Create a more realistic synthetic Titanic-like dataset
        np.random.seed(42)
        n_passengers = 891
        
        # Generate realistic passenger data
        # Age distribution (typical for 1912 ocean travel)
        ages = np.random.gamma(2, 15, n_passengers)  # Skewed toward younger passengers
        ages = np.clip(ages, 0.5, 80)
        
        # Class distribution (historical Titanic proportions)
        classes = np.random.choice([1, 2, 3], n_passengers, p=[0.24, 0.21, 0.55])
        
        # Sex distribution (more men than women historically)
        sexes = np.random.choice(['male', 'female'], n_passengers, p=[0.65, 0.35])
        
        # Generate family information
        sibsp = np.random.choice([0, 1, 2, 3, 4, 5], n_passengers, p=[0.68, 0.23, 0.06, 0.02, 0.01, 0.001])
        parch = np.random.choice([0, 1, 2, 3, 4, 5], n_passengers, p=[0.76, 0.13, 0.08, 0.02, 0.005, 0.005])
        
        # Fare based on class and family size
        base_fares = {1: 84, 2: 21, 3: 13}  # Average fares by class
        fares = np.array([base_fares[cls] * (1 + np.random.normal(0, 0.5)) * (1 + (s + p) * 0.1) 
                         for cls, s, p in zip(classes, sibsp, parch)])
        fares = np.clip(fares, 0, 500)
        
        # Embarked port
        embarked = np.random.choice(['S', 'C', 'Q'], n_passengers, p=[0.72, 0.19, 0.09])
        
        # Survival based on historical patterns
        # Women and children first, first class advantage
        survival_prob = np.zeros(n_passengers)
        
        for i in range(n_passengers):
            prob = 0.1  # Base survival rate
            
            # Women and children first
            if sexes[i] == 'female':
                prob += 0.6
            if ages[i] < 16:
                prob += 0.3
                
            # Class advantage
            if classes[i] == 1:
                prob += 0.3
            elif classes[i] == 2:
                prob += 0.1
            else:  # Third class
                prob -= 0.1
                
            # Family effects (very large families had worse survival)
            family_size = sibsp[i] + parch[i] + 1
            if family_size > 4:
                prob -= 0.2
            elif family_size == 1:  # Alone
                prob -= 0.1
                
            survival_prob[i] = np.clip(prob, 0, 1)
        
        survived = np.random.binomial(1, survival_prob)
        
        # Create DataFrame
        titanic = pd.DataFrame({
            'survived': survived,
            'pclass': classes,
            'sex': sexes,
            'age': ages,
            'sibsp': sibsp,
            'parch': parch,
            'fare': fares,
            'embarked': embarked
        })
        
        # Add some missing values to make it realistic (Age was often missing)
        missing_age_indices = np.random.choice(n_passengers, 177, replace=False)  # Historical missing count
        titanic.loc[missing_age_indices, 'age'] = np.nan
        
        # A few missing embarked values
        missing_embarked_indices = np.random.choice(n_passengers, 2, replace=False)
        titanic.loc[missing_embarked_indices, 'embarked'] = np.nan
        
        print(f"Synthetic Titanic dataset created: {titanic.shape}")
        return titanic

# Load Titanic data
titanic_data = load_titanic_dataset()

print("Titanic Dataset Overview:")
print(titanic_data.info())
print("\nTarget distribution:")
print(titanic_data['survived'].value_counts())

# Data exploration
print("\nBasic statistics:")
print(titanic_data.describe())

print("\nSurvival by key factors:")
if 'sex' in titanic_data.columns:
    print("By Sex:")
    print(titanic_data.groupby('sex')['survived'].agg(['count', 'mean']))

if 'pclass' in titanic_data.columns:
    print("\nBy Class:")
    print(titanic_data.groupby('pclass')['survived'].agg(['count', 'mean']))

# =============================================================================
# Baseline AutoGluon Model on Titanic
# =============================================================================

print("\n" + "="*30)
print("BASELINE AUTOGLUON MODEL")
print("="*30)

# Split Titanic data
titanic_train, titanic_test = train_test_split(
    titanic_data, 
    test_size=0.2, 
    random_state=42, 
    stratify=titanic_data['survived']
)

print(f"Titanic training set: {titanic_train.shape}")
print(f"Titanic test set: {titanic_test.shape}")

# Baseline model
titanic_predictor = TabularPredictor(
    label='survived',
    eval_metric='roc_auc',
    path='./titanic_baseline'
)

titanic_predictor.fit(titanic_train, time_limit=120)

# Evaluate baseline
titanic_predictions = titanic_predictor.predict(titanic_test)
titanic_accuracy = accuracy_score(titanic_test['survived'], titanic_predictions)
print(f"Baseline Titanic Accuracy: {titanic_accuracy:.4f}")

# Feature importance for Titanic
titanic_importance = titanic_predictor.feature_importance(titanic_test)
print("\nTitanic Feature Importance:")
print(titanic_importance)

# =============================================================================
# Custom Feature Engineering for Titanic
# =============================================================================

print("\n" + "="*30)
print("CUSTOM FEATURE ENGINEERING FOR TITANIC")
print("="*30)

def engineer_titanic_features(df):
    """Engineer domain-specific features for Titanic dataset"""
    df = df.copy()
    
    # Family size features
    if 'sibsp' in df.columns and 'parch' in df.columns:
        df['family_size'] = df['sibsp'] + df['parch'] + 1
        df['is_alone'] = (df['family_size'] == 1).astype(int)
    
    # Age groups
    if 'age' in df.columns:
        df['age_group'] = pd.cut(df['age'], 
                                bins=[0, 12, 20, 40, 60, 100], 
                                labels=['child', 'teen', 'adult', 'middle_age', 'senior'])
    
    # Fare per person
    if 'fare' in df.columns and 'family_size' in df.columns:
        df['fare_per_person'] = df['fare'] / df['family_size']
    
    # Title extraction (if name column exists)
    # This is a placeholder since our synthetic data doesn't have names
    if 'name' in df.columns:
        df['title'] = df['name'].str.extract(' ([A-Za-z]+)\.', expand=False)
        df['title'] = df['title'].replace(['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 
                                          'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
        df['title'] = df['title'].replace('Mlle', 'Miss')
        df['title'] = df['title'].replace('Ms', 'Miss')
        df['title'] = df['title'].replace('Mme', 'Mrs')
    
    # Class and sex interaction
    if 'pclass' in df.columns and 'sex' in df.columns:
        df['class_sex'] = df['pclass'].astype(str) + '_' + df['sex']
    
    return df

# Apply feature engineering
titanic_train_eng = engineer_titanic_features(titanic_train)
titanic_test_eng = engineer_titanic_features(titanic_test)

print("New features created:")
original_cols = set(titanic_train.columns)
new_cols = set(titanic_train_eng.columns) - original_cols
for col in new_cols:
    print(f"- {col}")

# Train improved model
titanic_improved_predictor = TabularPredictor(
    label='survived',
    eval_metric='roc_auc',
    path='./titanic_improved'
)

titanic_improved_predictor.fit(titanic_train_eng, time_limit=120)

# Evaluate improved model
titanic_improved_pred = titanic_improved_predictor.predict(titanic_test_eng)
titanic_improved_accuracy = accuracy_score(titanic_test_eng['survived'], titanic_improved_pred)

print(f"\nBaseline Accuracy: {titanic_accuracy:.4f}")
print(f"Improved Accuracy: {titanic_improved_accuracy:.4f}")
print(f"Improvement: {titanic_improved_accuracy - titanic_accuracy:.4f}")

# =============================================================================
# Model Interpretation for Titanic
# =============================================================================

print("\n" + "="*30)
print("TITANIC MODEL INTERPRETATION")
print("="*30)

# Feature importance comparison
baseline_importance = titanic_predictor.feature_importance(titanic_test)
improved_importance = titanic_improved_predictor.feature_importance(titanic_test_eng)

print("Baseline Model - Top Features:")
print(baseline_importance.head())

print("\nImproved Model - Top Features:")
print(improved_importance.head())

# Model leaderboard
print("\nImproved Model Leaderboard:")
titanic_leaderboard = titanic_improved_predictor.leaderboard(titanic_test_eng, silent=True)
print(titanic_leaderboard.head())

# =============================================================================
# Performance Evaluation and Comparison
# =============================================================================

print("\n" + "="*30)
print("COMPREHENSIVE PERFORMANCE EVALUATION")
print("="*30)

# Compare multiple approaches
results_comparison = []

# AutoGluon baseline
results_comparison.append({
    'Model': 'AutoGluon Baseline',
    'Accuracy': titanic_accuracy,
    'Features': len(titanic_train.columns) - 1  # Exclude target
})

# AutoGluon improved
results_comparison.append({
    'Model': 'AutoGluon Improved',
    'Accuracy': titanic_improved_accuracy,
    'Features': len(titanic_train_eng.columns) - 1  # Exclude target
})

# Simple baseline (majority class)
majority_accuracy = titanic_test['survived'].value_counts().max() / len(titanic_test)
results_comparison.append({
    'Model': 'Majority Class Baseline',
    'Accuracy': majority_accuracy,
    'Features': 0
})

# Simple logistic regression for comparison
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

try:
    # Prepare data for sklearn
    numeric_features = titanic_train.select_dtypes(include=[np.number]).columns.tolist()
    if 'survived' in numeric_features:
        numeric_features.remove('survived')

    categorical_features = titanic_train.select_dtypes(include=['object']).columns.tolist()

    # Create preprocessing pipeline with imputation
    numeric_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ])

    categorical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
        ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore'))
    ])

    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numeric_transformer, numeric_features),
            ('cat', categorical_transformer, categorical_features)
        ])

    # Simple logistic regression
    lr_model = LogisticRegression(random_state=42, max_iter=1000)

    # Prepare data
    X_train_lr = titanic_train.drop('survived', axis=1)
    y_train_lr = titanic_train['survived']
    X_test_lr = titanic_test.drop('survived', axis=1)
    y_test_lr = titanic_test['survived']

    # Fit preprocessing and model
    X_train_processed = preprocessor.fit_transform(X_train_lr)
    X_test_processed = preprocessor.transform(X_test_lr)

    lr_model.fit(X_train_processed, y_train_lr)
    lr_predictions = lr_model.predict(X_test_processed)
    lr_accuracy = accuracy_score(y_test_lr, lr_predictions)

    results_comparison.append({
        'Model': 'Logistic Regression (with preprocessing)',
        'Accuracy': lr_accuracy,
        'Features': X_train_processed.shape[1]
    })
    
    print(f"Logistic regression (with preprocessing) accuracy: {lr_accuracy:.4f}")

except Exception as e:
    print(f"Could not train logistic regression: {e}")
    # Add a placeholder result
    results_comparison.append({
        'Model': 'Logistic Regression (failed)',
        'Accuracy': 0.0,
        'Features': 0
    })

# Display comparison
comparison_df = pd.DataFrame(results_comparison)
comparison_df = comparison_df.sort_values('Accuracy', ascending=False)
print("Model Comparison:")
print(comparison_df)

# =============================================================================
# Model Deployment Preparation
# =============================================================================

print("\n" + "="*30)
print("MODEL DEPLOYMENT PREPARATION")
print("="*30)

# Save the best model
best_model_path = './titanic_final_model'
final_predictor = titanic_improved_predictor

# The model is already saved, but let's demonstrate loading
def load_saved_model(path):
    """Load a saved AutoGluon model"""
    loaded_predictor = TabularPredictor.load(path)
    return loaded_predictor

# Create prediction function for deployment
def predict_survival(passenger_data):
    """
    Predict survival probability for new passengers
    
    Args:
        passenger_data: dict or DataFrame with passenger information
    
    Returns:
        dict with prediction and probability
    """
    if isinstance(passenger_data, dict):
        passenger_df = pd.DataFrame([passenger_data])
    else:
        passenger_df = passenger_data.copy()
    
    # Apply same feature engineering
    passenger_df = engineer_titanic_features(passenger_df)
    
    # Get the columns that the model was trained on
    try:
        # Try to get the training columns from the predictor
        training_columns = final_predictor.feature_metadata_in.get_features()
    except:
        # Fallback: use the columns from our training data (excluding target)
        training_columns = [col for col in titanic_train_eng.columns if col != 'survived']
    
    # Add any missing columns with default values
    for col in training_columns:
        if col not in passenger_df.columns:
            # Add missing columns with appropriate default values
            if col in ['class', 'who', 'adult_male', 'deck', 'alive', 'alone']:
                # These are seaborn-specific derived columns, provide defaults
                if col == 'class':
                    passenger_df[col] = 'Third'  # Default class
                elif col == 'who':
                    passenger_df[col] = 'man' if passenger_df['sex'].iloc[0] == 'male' else 'woman'
                elif col == 'adult_male':
                    passenger_df[col] = (passenger_df['sex'] == 'male') & (passenger_df['age'] >= 18)
                elif col == 'deck':
                    passenger_df[col] = 'Unknown'  # Most deck info was missing
                elif col == 'alive':
                    passenger_df[col] = 'unknown'  # This would be what we're predicting
                elif col == 'alone':
                    passenger_df[col] = passenger_df.get('is_alone', 1)  # Use our engineered feature if available
            else:
                # For other missing columns, use appropriate defaults
                if passenger_df[col].dtype == 'object':
                    passenger_df[col] = 'Unknown'
                else:
                    passenger_df[col] = 0
    
    # Ensure we only use the columns the model expects
    passenger_df = passenger_df[training_columns]
    
    try:
        # Make prediction
        prediction = final_predictor.predict(passenger_df)
        probability = final_predictor.predict_proba(passenger_df)
        
        return {
            'prediction': prediction.iloc[0] if hasattr(prediction, 'iloc') else prediction[0],
            'survival_probability': probability.iloc[0, 1] if hasattr(probability, 'iloc') else probability[0][1]
        }
    except Exception as e:
        print(f"Prediction failed: {e}")
        print(f"Required columns: {training_columns}")
        print(f"Provided columns: {list(passenger_df.columns)}")
        return {
            'prediction': None,
            'survival_probability': None,
            'error': str(e)
        }

# Test the prediction function
sample_passenger = {
    'pclass': 1,
    'sex': 'female',
    'age': 25,
    'sibsp': 1,
    'parch': 0,
    'fare': 50,
    'embarked': 'S'
}

result = predict_survival(sample_passenger)
print(f"Sample prediction:")
print(f"Passenger: {sample_passenger}")
print(f"Survival prediction: {result['prediction']}")
print(f"Survival probability: {result['survival_probability']:.3f}")

# =============================================================================
# PROJECT SUMMARY AND BUSINESS IMPACT
# =============================================================================

print("\n" + "="*50)
print("PROJECT SUMMARY AND BUSINESS IMPACT")
print("="*50)

print("Titanic Survival Prediction Project Summary:")
print("=" * 45)

print(f"1. Dataset: {len(titanic_data)} passengers")
print(f"2. Baseline accuracy: {titanic_accuracy:.3f}")
print(f"3. Improved accuracy: {titanic_improved_accuracy:.3f}")
print(f"4. Improvement: {(titanic_improved_accuracy - titanic_accuracy):.3f}")

print(f"\nKey Success Factors:")
print(f"- Automatic preprocessing handled missing values effectively")
print(f"- Feature engineering improved model performance")
print(f"- Ensemble methods provided robust predictions")
print(f"- Model interpretation revealed logical patterns")

print(f"\nBusiness Applications:")
print(f"- Risk assessment for insurance pricing")
print(f"- Safety protocol development")
print(f"- Emergency response planning")
print(f"- Historical analysis and education")

print(f"\nTechnical Achievements:")
print(f"- Rapid prototyping: Working model in minutes")
print(f"- Feature importance: Identified key survival factors")
print(f"- Model transparency: Explainable predictions")
print(f"- Production ready: Deployable prediction function")

print("\n" + "="*50)
print("CHAPTER 6 COMPLETE!")
print("="*50)
print("You've successfully learned to:")
print("1. Set up and use AutoGluon TabularPredictor")
print("2. Handle classification and regression tasks")
print("3. Leverage automatic data processing")
print("4. Customize models for specific requirements")
print("5. Build end-to-end ML projects")
print("6. Prepare models for deployment")
```
